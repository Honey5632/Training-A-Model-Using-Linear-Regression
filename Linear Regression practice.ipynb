```python
# importing library

import pandas as pd
import numpy as np
import matplotlib as plt
from sklearn.datasets._base import load_diabetes
```

```python
# loading predefined dataset from sklearn 
df = load_diabetes()
df
```

```python
# now creating a dataframe for the given dataset
dataset = pd.DataFrame(df.data)
dataset
```

```python
# adding feature_names for table columns
dataset.columns = df.feature_names
```

```python
dataset
```

```python
# first 5 elements
dataset.head()
```

```python
# take variables for traning and testing
X = dataset
y = df.target
```

```python
y
```

```python
# implement traning and testing by using train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train =scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

```python
# fit to the linear regression
regression = LinearRegression()
regression.fit(X_train,y_train)
```

```python
# cross validator score
mse = cross_val_score(regression, X_train, y_train, scoring='neg_mean_squared_error', cv=10)
mse
```

```python
# finding the mean in cross view
np.mean(mse)
```

```python
# predicted
reg_pred = regression.predict(X_test)
```

```python
# predicted values
reg_pred
```

```python
import seaborn as sns
sns.displot(reg_pred-y_test,kind='kde')
```

```python
from sklearn.metrics import r2_score
score = r2_score(reg_pred,y_test)
```
